```python
from sklearn import datasets

cancer = datasets.load_breast_cancer()

print("Features: ", cancer.feature_names)

print("Labels: ", cancer.target_names)
```

```python
print(cancer.data.shape)
print(cancer.data[0:5])
```

```python
# print the cancer labels (0:malignant, 1:benign)
print(cancer.target)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3,random_state=109)
```

```python
from sklearn import svm

clf = svm.SVC(kernel='linear') # Linear Kernel

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to plot the dataset
def plot_data(X, y, title):
    plt.figure(figsize=(6, 4))
    plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='red', label='Class 0')
    plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', label='Class 1')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(title)
    plt.legend()
    plt.show()

# Read the datasets
linear_data = pd.read_csv("linear_kernel_data.csv")
rbf_data = pd.read_csv("rbf_kernel_data.csv")
poly_data = pd.read_csv("poly_kernel_data.csv")

# Extract features (X) and labels (y)
X_linear = linear_data[['Feature1', 'Feature2']].values
y_linear = linear_data['Target'].values

X_rbf = rbf_data[['Feature1', 'Feature2']].values
y_rbf = rbf_data['Target'].values

X_poly = poly_data[['Feature1', 'Feature2']].values
y_poly = poly_data['Target'].values

# Plot the datasets
plot_data(X_linear, y_linear, "Linear Kernel Dataset")
plot_data(X_rbf, y_rbf, "RBF Kernel Dataset")
plot_data(X_poly, y_poly, "Polynomial Kernel Dataset")
```

```python
def plot_data(X, y, clf, title):
    plt.figure(figsize=(8, 6))

    # Plotting class 0
    plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='red', label='Class 0', edgecolors='k')

    # Plotting class 1
    plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', label='Class 1', edgecolors='k')

    # Create grid to plot decision boundary
    xx, yy = np.meshgrid(np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 100),
                         np.linspace(X[:, 1].min() - 1, X[:, 1].max() + 1, 100))

    # Predict class labels for each point on the grid
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot the decision boundary
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)

    # Adding labels and title
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
linear_data = pd.read_csv("linear_kernel_data.csv")
X_linear = linear_data[['Feature1', 'Feature2']].values
y_linear = linear_data['Target'].values

clf = svm.SVC(kernel='linear')
clf.fit(X_linear, y_linear)

# Plotting the data and decision boundary
plot_data(X_linear, y_linear, clf, 'SVM Classifier with Linear Kernel')
y_pred = clf.predict(X_linear)
print("Accuracy:",metrics.accuracy_score(y_linear, y_pred))
```
