```python
from sklearn.datasets import make_classification
X, y = make_classification(
    n_samples=1000,
    n_features=20,
    n_informative=10,
    n_redundant=5,
    n_repeated=0,
    n_classes=2,
    flip_y=0.1,  # bruit dans les labels
    class_sep=0.5,  # classes proches
    random_state=42
)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
# Affichage
plt.figure(figsize=(8, 6))
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=plt.cm.RdYlBu, edgecolor='k')
plt.xlabel("Composante principale 1")
plt.ylabel("Composante principale 2")
plt.grid(True)
plt.show()

# Affichage
plt.figure(figsize=(8, 6))
plt.scatter(X_train[:, 2], X_train[:, 3], c=y_train, cmap=plt.cm.RdYlBu, edgecolor='k')
plt.xlabel("Composante principale 2")
plt.ylabel("Composante principale 3")
plt.grid(True)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

tree_model = DecisionTreeClassifier(max_depth=10, random_state=42)
tree_model.fit(X_train, y_train)
y_pred_tree = tree_model.predict(X_test)
print("Accuracy (Decision Tree):", accuracy_score(y_test, y_pred_tree))
```

```python
from sklearn.ensemble import RandomForestClassifier
forest_model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
forest_model.fit(X_train, y_train)
y_pred_forest = forest_model.predict(X_test)
print("Accuracy (Random Forest):", accuracy_score(y_test, y_pred_forest))
```

```python
X_bruit, y_bruit = make_classification(
    n_samples=1000,
    n_features=20,
    n_informative=10,
    n_redundant=5,
    n_repeated=0,
    n_classes=2,
    flip_y=0.1,  # bruit dans les labels
    class_sep=0.5,  # classes proches
    random_state=42
)

X_train, X_test, y_train, y_test = train_test_split(X_bruit, y_bruit, test_size=0.3, random_state=42)

tree_model = DecisionTreeClassifier(random_state=42)
forest_model = RandomForestClassifier(n_estimators=100, random_state=42)

tree_model.fit(X_train, y_train)
forest_model.fit(X_train, y_train)

y_pred_tree = tree_model.predict(X_test)
y_pred_forest = forest_model.predict(X_test)

print("Accuracy (Decision Tree):", accuracy_score(y_test, y_pred_tree))
print("Accuracy (Random Forest):", accuracy_score(y_test, y_pred_forest))
```
