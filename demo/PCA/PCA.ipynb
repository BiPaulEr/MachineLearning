```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.datasets import make_classification

def plot(X, y, title):
    plt.figure(figsize=(6, 5))
    h = 0.02  # step size in the mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
    plt.title(title)
    plt.show()
    
X_1, y_1 = make_classification(n_samples=200, n_features=6, n_redundant=2, n_informative=3,
                                         n_clusters_per_class=1, class_sep=2, random_state=42)
plot(X_1, y_1, "dataset")
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_1)
plot(X_scaled, y_1, "scaled dataset")
```

```python
pca = PCA(n_components=4)
X_pca = pca.fit_transform(X_scaled)

plot(X_pca, y_1, "scaled dataset")
```

```python
print("Explained variance:", pca.explained_variance_ratio_)
print("Cumulative:", np.cumsum(pca.explained_variance_ratio_))
```
