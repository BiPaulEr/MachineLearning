**Demonstration que SSE ne resoud pas tout**

```python
def test_kmeans_inertia(data, max_k):
    inertias = []
    for k in range(1, max_k + 1):
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(data)
        inertias.append(kmeans.inertia_)
    return inertias
```

```python
S = create_data(4, 1.5)
inertias = test_kmeans_inertia(S, max_k=8)
```

```python
def plot_elbow_curve(inertias):
    ks = range(1, len(inertias) + 1)
    plt.plot(ks, inertias, marker='o')
    plt.title("Mthode du coude (Elbow Method)")
    plt.xlabel("Nombre de clusters (k)")
    plt.ylabel("SSE (Inertie)")
    plt.grid(True)
    plt.show()

plot_elbow_curve(inertias)
```
