```python
from sklearn.datasets import make_blobs

def create_data(centers, cluster_std=0.6):
    X, _ = make_blobs(n_samples=150, centers=centers, cluster_std=cluster_std, random_state=0)
    return X
X = create_data(4)
print(X[:5])
```

```python
import matplotlib.pyplot as plt
import numpy as np
def plot_clusters(X, labels, title):
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow')
    plt.title(title)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.grid(True)
    plt.show()
plot_clusters(X, np.zeros(X.shape[0]), "data")
```

```python
from sklearn.cluster import DBSCAN

def dbscan_clustering(X, eps=0.5, min_samples=5):
    model = DBSCAN(eps=eps, min_samples=min_samples)
    labels = model.fit_predict(X)
    return labels
labels = dbscan_clustering(X, 0.2, 5)
plot_clusters(X, labels, "4 centers eps 0.2 samples 5")
labels = dbscan_clustering(X, 0.5, 5)
plot_clusters(X, labels, "4 centers eps 0.5 samples 5")
labels = dbscan_clustering(X, 0.8, 5)
plot_clusters(X, labels, "4 centers eps 0.8 samples 5")
```

```python
Y = create_data(8)

labels_Y = dbscan_clustering(Y, 0.5, 5)
plot_clusters(Y , labels_Y, "8 centers eps 0.5 samples 5")

labels_Y = dbscan_clustering(Y, 0.8, 5)
plot_clusters(Y , labels_Y, "8 centers eps 0.8 samples 5")

labels_Y = dbscan_clustering(Y, 1, 5)
plot_clusters(Y , labels_Y, "8 centers eps 1 samples 5")
```

```python
Z = create_data(4, 2)

labels_Z = dbscan_clustering(Z, 0.5, 5)
plot_clusters(Z , labels_Z, "8 centers eps 0.5 samples 5")

labels_Z = dbscan_clustering(Z, 0.8, 5)
plot_clusters(Z, labels_Z, "8 centers eps 0.8 samples 5")

labels_Z = dbscan_clustering(Z, 1, 5)
plot_clusters(Z , labels_Z, " centers eps 1 samples 5")
```
