```python
from sklearn.datasets import make_blobs

def create_data(centers, cluster_std=0.6):
    X, _ = make_blobs(n_samples=150, centers=centers, cluster_std=cluster_std, random_state=1)
    return X
X = create_data(4)
print(X[:5])
```

```python
import matplotlib.pyplot as plt
import numpy as np
def plot_clusters(X, labels, title):
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow')
    plt.title(title)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.grid(True)
    plt.show()
plot_clusters(X, np.zeros(X.shape[0]), "data")
```

```python
from sklearn.cluster import AgglomerativeClustering

def agglomerative_clustering(X, n_clusters=4):
    model = AgglomerativeClustering(n_clusters=n_clusters)
    labels = model.fit_predict(X)
    return labels
labels = agglomerative_clustering(X, 4)
plot_clusters(X, labels, "4 centers")
```

```python
Y = create_data(8)
labels_Y = agglomerative_clustering(Y, 8)
plot_clusters(Y, labels_Y, "8 centers")
```

```python
Z = create_data(4, 2)
labels_Z = agglomerative_clustering(Z, 4)
plot_clusters(Z, labels_Z, "4 centers with high variance")
```
